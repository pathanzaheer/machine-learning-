1.Basic Line Plot
o Plot a simple line graph of ùë¶ = ùë•2 for ùë• values ranging from -10 to 10.
o Label the x-axis as "X Values" and the y-axis as "Squared Values."
o Add a title to the graph.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 11)
y = x**2

# Plot the graph
plt.plot(x, y)
plt.xlabel("X Values")
plt.ylabel("Squared Values")
plt.title("Line Plot of y = x^2")
plt.grid(True)
plt.show()
```

2.Bar Plot
o Create a bar chart showing the marks scored by 5 students: [Sai, 'Vamsi', 'Divij', 'Sritan', 'Sita'] with marks [85, 90, 78, 92, 88].
o Add labels to the x-axis and y-axis and give the chart a title.

```python

# Data
students = ['Sai', 'Vamsi', 'Divij', 'Sritan', 'Sita']
marks = [85, 90, 78, 92, 88]

# Plot the bar chart
plt.bar(students, marks, color='skyblue')
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Marks of Students")
plt.show()
```

3.Scatter Plot
o Generate two lists of 20 random integers each (for x and y coordinates).
o Create a scatter plot with these values, and customize it by adding color, size, and labels to the points.

```python
# Generate random data
x = np.random.randint(1, 100, 20)
y = np.random.randint(1, 100, 20)
sizes = np.random.randint(20, 200, 20)
colors = np.random.rand(20)

# Plot scatter plot
plt.scatter(x, y, s=sizes, c=colors, alpha=0.7, cmap='viridis')
plt.colorbar(label="Color Scale")
plt.xlabel("X Coordinates")
plt.ylabel("Y Coordinates")
plt.title("Scatter Plot with Customizations")
plt.show()
```

4.Histogram
o Simulate 100 random values from a normal distribution using np.random.randn() and plot a histogram.
o Customize the number of bins and add a grid to the plot.

```python
# Generate random data
random_values = np.random.randn(100)

# Plot histogram
plt.hist(random_values, bins=15, color='purple', edgecolor='black')
plt.grid(True)
plt.title("Histogram of Random Values")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

5.Pie Chart
o Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E'] with values [20, 30, 25, 15, 10].
o Display the percentage values on the chart and use a color palette for differentiation.

```python

# Data
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]

# Plot pie chart
plt.pie(values, labels=categories, autopct="%.1f%%", colors=plt.cm.Paired.colors, startangle=90)
plt.title("Percentage Distribution of Categories")
plt.show()
```

6.Box Plot (Seaborn)
o Create a box plot to visualize the distribution of the marks scored by students in a class: [78, 85, 92, 88, 76, 95, 89, 84, 91, 73].
o Highlight the median and any outliers.

```python
import seaborn as sns

# Data
marks = [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]

# Plot box plot
sns.boxplot(data=marks, color="lightblue")
plt.title("Box Plot of Student Marks")
plt.show()
```

7.Heatmap (Seaborn)
o Create a 5x5 matrix with random integers between 1 and 10 using np.random.randint() and visualize it as a heatmap using Seaborn.
o Add a color bar and annotate the heatmap with the values.

```python
# Generate random matrix
data = np.random.randint(1, 11, size=(5, 5))

# Plot heatmap
sns.heatmap(data, annot=True, cmap="coolwarm", cbar=True)
plt.title("Heatmap of Random Matrix")
plt.show()
```

8.Comparison of Matplotlib and Seaborn
o Generate a pair of plots (bar charts or line plots) using the same data in both Matplotlib and Seaborn.
o Highlight the differences in the code, syntax, and aesthetics of the two libraries.

```python
# Data
x = ['A', 'B', 'C', 'D', 'E']
y = [5, 7, 8, 6, 9]

# Matplotlib bar plot
plt.bar(x, y, color='orange')
plt.title("Matplotlib Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()

# Seaborn bar plot
sns.barplot(x=x, y=y, palette="Blues")
plt.title("Seaborn Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

10.Save and Style Plots
‚Ä¢ Plot any graph of your choice and save it in both .png and .pdf formats.
‚Ä¢ Apply at least one built-in style (e.g., 'ggplot' or 'seaborn-darkgrid') and explain how the style impacts the visualization.

```python
# Apply style
plt.style.use('ggplot')

# Plot line graph
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label="Sine Wave")
plt.title("Styled Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()

# Save the plot
plt.savefig("styled_plot.png")
plt.savefig("styled_plot.pdf")
plt.show()
```
