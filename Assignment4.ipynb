# **EDA on Most Visited Countries in 2024**

*   **Column Descriptions:**

1.   country: Names of countries (203 unique entries, non-null).
2.   MostVisited_NumOfArrivalsPredictive_Millions_2024: Predicted     
     number of arrivals in 2024 (50 non-null, float).
3.   MostVisited_NumOfArrivals_Millions_2023: Number of arrivals in   
     2023 (49 non-null, float).
4.   MostVisited_NumOfArrivals_Millions_2022: Number of arrivals in
     2022 (20 non-null, float).
5.   MostVisited_NumOfArrivals_WorldBank: Historical number of  
     arrivals (202 non-null, float).
6.   MostVisited_DataYear_WorldBank: Year of data from World Bank (202
non-null, float).

   




```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.set(style="whitegrid")
data=pd.read_csv("/content/most-visited-countries-2024.csv")
```

*   **Predicted Arrivals for 2024:**

1.   The distribution is slightly right-skewed with most predictions clustered below 30 million arrivals.
2.   A small number of countries have significantly higher predicted arrivals.



```python
plt.figure(figsize=(10, 6))
sns.histplot(data['MostVisited_NumOfArrivalsPredictive_Millions_2024'], bins=15, kde=True, color='skyblue')
plt.title('Distribution of Predicted Arrivals (Millions, 2024)', fontsize=16)
plt.xlabel('Predicted Arrivals (Millions)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```





*   **Missing Data Heatmap**:


1.   There is a significant amount of missing data in the 2023 and 2022 arrival columns, with better coverage for the World Bank historical data.




```python
plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing Data Heatmap', fontsize=16)
plt.show()
```

*   **Top 10 Countries by Historical Arrivals:**


1.   The top countries exhibit a steep disparity in the number of arrivals, with a few significantly outperforming others based on World Bank data.



```python
top_countries = data.nlargest(10, 'MostVisited_NumOfArrivals_WorldBank')

plt.figure(figsize=(12, 8))
sns.barplot(
    x='MostVisited_NumOfArrivals_WorldBank',
    y='country',
    data=top_countries,
    palette='viridis'
)
plt.title('Top 10 Countries by Historical Arrivals (WorldBank Data)', fontsize=16)
plt.xlabel('Number of Arrivals (World Bank)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.show()
```
